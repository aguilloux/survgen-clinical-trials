```python
import os
import pandas as pd

import sys
from pathlib import Path
module_path = Path.cwd().parent / 'utils'
sys.path.append(str(module_path))
from visualization import visualize_general_perf

import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
def visualize(dataset_name):
    current_path = os.getcwd()  # Get current working directory
    parent_path = os.path.dirname(current_path)
    res_dir = parent_path + "/dataset/" + dataset_name + "/metric_results/"

    general_scores_df_ext = pd.read_csv(res_dir + "general_scores_df_ext.csv")
    general_scores_df = pd.read_csv(res_dir + "general_scores_df.csv")
    visualize_general_perf(general_scores_df_ext, metrics=[['J-S distance', "min"], ['KS test', "max"]], title="Train on control+treated")
    visualize_general_perf(general_scores_df, metrics=[['J-S distance', "min"], ['KS test', "max"]], title="Train on only control")
    visualize_general_perf(general_scores_df_ext, metrics=[['Survival curves distance', "min"], ['Detection XGB', "min"]], title="Train on control+treated")
    visualize_general_perf(general_scores_df, metrics=[['Survival curves distance', "min"], ['Detection XGB', "min"]], title="Train on only control")
    visualize_general_perf(general_scores_df_ext, metrics=[['NNDR', "max"], ['K-map score', "max"]], title="Train on control+treated")
    visualize_general_perf(general_scores_df, metrics=[['NNDR', "max"], ['K-map score', "max"]], title="Train on only control")
```

# Aids
